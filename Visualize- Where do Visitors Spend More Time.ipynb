### picking one particular date

```python
temp_df = df.loc[df.just_date=="2018-06-16"]
```

#### Extract Latitude, longitude ponits

```python

lat_list =[]
lon_list = []
def extract_latlng(row):
    lat_list.extend(eval(row['lat']))
    lon_list.extend(eval(row['lng']))
```

```python
temp_df[["lat","lng"]].apply(lambda row: extract_latlng(row),axis=1)
```

```python
d = {'lat':lat_list,'lng':lon_list}
df2 = pd.DataFrame(d)
lat_lon = df2.values.tolist()
```

### Plotting movement of a single mac_id

```python
loc_info = literal_eval(df.iloc[1001,2])
```

```python
loc_info = sorted(loc_info, key=lambda e: e[-1], reverse=False)
lat_lon = [[i[1],i[2],1] for i in loc_info]
len(lat_lon)
```

```python
my_data =[[i] for i in lat_lon]
m = folium.Map(my_data[0][0][0:2], tiles='OpenStreetMap', zoom_start=25)
hm = plugins.HeatMapWithTime(my_data,auto_play=True,max_speed=20,speed_step=1,position="topleft")
hm.add_to(m)
m
```
